**Respuesta**

Para normalizar $p(x|\omega_i)$ debemos hallar una constante de normalización N tal que 

$$p(x|\omega_i) = Ne^{(-|x-a_i|/b_i)}$$ y $$\int Ne^{(-|x-a_i|/b_i)}dx = 1$$


**prueba**

* (b) Calcule el radio de verosimilitud como funcion de los parametros.

$$p(x|\omega_1)/p(x|\omega_2)$$ 

=
$$\frac {\frac{1}{2b1} e^{(-|x-a_1|/b_1)}}{\frac{1}{2b2} e^{(-|x-a_2|/b_2)}}$$


=
$$\frac {\frac{e^{(-|x-a_1|/b_1)}}{2b1}}{\frac{e^{(-|x-a_2|/b_2)}}{2b2}}$$


=
$$\frac { 2b2 e^{(-|x-a_1|/b_1)}} {2b1 e^{(-|x-a_2|/b_2)}}$$

$$\frac { b2}{b1} e^{(-|x-a_1|/b_1 + |x-a_2|/b_2)}$$


* (c) Graficar el radio $p(x|\omega_1)/p(x|\omega_2)$ para el caso $a_1=0$, $b_1=1$, $a_2=1$ y $b_2=2$.


```python
import numpy as np
import matplotlib.pyplot as plt

# p(x|wi)
def verosimilitud_normalizada(x, a, b):
    N = 1./2*b
    return N * np.exp(-np.abs(x - a)/b)

x = np.linspace(-10.0, 10.0)
y1 = verosimilitud_normalizada(x, 0, 1)
y2 = verosimilitud_normalizada(x, 1, 2)

y = y1/y2

plt.plot(x, y, 'o-')
plt.title('Razon de verosimilitud')
plt.show()
```

**Respuesta**

La grafica obtenida en el inciso b) demuestra que para los parametros dados: $a_1=3$, $a_2=5$ y $b = 1$, $P (\omega_1 | x)$ converge a 0.5 en ambos casos. i.e.:

 $$\lim_{x \to -\infty} (\omega_1 | x) = \lim_{x \to \infty}(\omega_1 | x)  = 0.5 $$
 
En el caso de $P (\omega_2 | x)$ tambien se verifica que converge a 0.5 en ambos casos:

  $$\lim_{x \to -\infty} (\omega_2 | x) = \lim_{x \to \infty}(\omega_2 | x)  = 0.5 $$

```python
import numpy as np
import matplotlib.pyplot as plt

# p(x|wi)
def cauchy(x, a, b):
    return (1./np.pi*b) * 1./(1+ ((x - a)/b)**2)
    
# p(w1|x)
def posteriori_w1(x, a1, a2, b):
    priori = 0.5
    evidencia = cauchy(x, a1, b)*priori + cauchy(x, a2, b)*priori
    return cauchy(x, a1, b)* priori/evidencia
                      

# p(w2|x)
def posteriori_w2(x, a1, a2, b):
    priori = 0.5
    evidencia = cauchy(x, a1, b)*priori + cauchy(x, a2, b)*priori
    return cauchy(x, a2, b)* priori/evidencia    
x = np.linspace(-10000.0, 10000.0)

a1 = 3
a2 = 5
b = 1

y = posteriori_w1(x, a1, a2, b)
plt.plot(x, y, 'o-')
plt.title('Probabilidad Posteriori para P(w1|x)')
plt.show()

y = posteriori_w2(x, a1, a2, b)
plt.plot(x, y, 'ro-')
plt.title('Probabilidad Posteriori para P(w2|x)')

plt.show()
```
