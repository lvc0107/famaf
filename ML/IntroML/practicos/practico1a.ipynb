Ejercicio 1: Siguiendo el código anterior, realice ``todos`` los gráficos de dos variables, cargando todos los 150 datos con sus correspondientes etiquetas. Incruste todos los gráficos en una sola figura.

```python
import matplotlib.pyplot as plt
import pandas as pd

iris_data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(iris_data_url)

def plot_data(data, i):
    fig.subplots_adjust(hspace=1.4, wspace=0.4)
    plt.subplot(1, 2, i)
    plt.scatter(data[:50, 0], data[:50, 1], color='red', marker='o', label='setosa')
    plt.scatter(data[50:100, 0], data[50:100, 1], color='blue', marker='x', label='versicolor')
    plt.scatter(data[100:150, 0], data[100:150, 1], color='green', marker='+', label='virginica')
    plt.legend(loc='upper left')

fig = plt.figure()
# extract sepal length and sepal width
data = df.iloc[0:150, [0, 1]].values
plot_data(data, 1)
plt.xlabel('sepal length [cm]')
plt.ylabel('sepal width [cm]')

# extract petal length and petal width
data = df.iloc[0:150, [2, 3]].values
plot_data(data, 2)
plt.xlabel('petal length [cm]')
plt.ylabel('petal width [cm]')

fig = plt.figure()
# extract sepal length and petal length
data = df.iloc[0:150, [0, 2]].values
plot_data(data, 1)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')

# extract sepal width and petal width
data = df.iloc[0:150, [1, 3]].values
plot_data(data, 2)
plt.xlabel('sepal width [cm]')
plt.ylabel('setal width [cm]')

fig = plt.figure()
# extract sepal length and petal width
data = df.iloc[0:150, [0, 3]].values
plot_data(data, 1)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal width [cm]')

# extract sepal width and petal length
data = df.iloc[0:150, [1, 2]].values
plot_data(data, 2)
plt.xlabel('sepal width [cm]')
plt.ylabel('setal length [cm]')
plt.show()
```
